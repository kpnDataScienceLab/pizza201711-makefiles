## 1. Problem description 

1. In folder books we have 4 different txt files (abyss.txt, isles.txt, last.txt, sierra.txt), which are obtained from the following sources respectively (see md file in books folder for more details):

    * [The People of the Abyss by Jack London](http://www.gutenberg.org/ebooks/1688)
    * [A Journey to the Western Islands of Scotland by Samuel Johnson](http://www.gutenberg.org/cache/epub/2064/pg2064.txt)
    * [Scott's Last Expedition Volume I by Robert Falcon Scott](http://www.gutenberg.org/ebooks/11579)
    * [My First Summer in the Sierra by John Muir](http://www.gutenberg.org/cache/epub/32540/pg32540.txt)
<br><br>
2. For each book we want to test [the Zipf´s law for NLP](https://en.wikipedia.org/wiki/Zipf%27s_law) that states that the frequency of any word is inversely proportional to its rank in the frequency table. To do that, we will:
    * run **wordcount.py** script: it will load a txt file, calculate the frequencies of each word in the file and
    save in a new file the words, counts and percentages of the total in descending order
    
            - Dependency: .txt file
            - Target: .dat file
        
    * run **plotcount.py** script: it will use .dat file created before and will generate a plot with word frequency distribution
        
            - Dependency: .dat file
            - Target: .png file
    * run **zipf_test.py** script: it will use .dat file created before and will return the top two word counts from .dat file. Results of test should be added to results.txt file
        
            - Dependency: .dat file
            - Target: result.txt file
            
    * results.txt file will contain the top two word counts per book

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import glob
%matplotlib inline

images = []
image_files = glob.glob('pngs/*.png')
for img_path in image_files:
    images.append(mpimg.imread(img_path))

plt.figure(figsize=(20,30),dpi=200)
columns = 2
for i, image in enumerate(images):
    plt.subplot(len(images) / columns + 1, columns, i + 1)
    plt.axis('off')
    plt.imshow(image)
    plt.title(image_files[i].split('/')[1])
```

### Drawbacks: example###

What if we have changed plotcount.py script (for example, adjusted plot parameters)? We do not need to rerun wordcount.py and zipf_test.py scripts again since the targets there will not change.

If we want to keep using the sh script, we will have to outcomment some lines or create some if statements. Not handy, especially if we have a big complicated shell script where many things may change!

### 4.2. Makefile is not pretty! Prettify it :-)

#### 4.2.1. D.R.Y. (Don’t Repeat Yourself) rule:

- Use \$@ to refer to the target of the current rule.
- Use \$^ to refer to the dependencies of the current rule.
- Use \$< to refer to the first dependency of the current rule.
   
#### 4.2.2. Add script as dependency
Makes sense the target is product of both input file and script -> in that case, if script changes, the part in makefile will rerun (otherwise it will assume nothing has changed)
<br><br>
Per default, make looks for a file called Makefile. But Makefile does not have to be called Makefile -> the pretty one after steps 1 and 2  has name Pretty_Makefile1

```python
!cat Pretty_Makefile1
```

```python
!make -f Pretty_Makefile1
```

```python
!make -f Pretty_Makefile1 clean
```

#### 4.2.3. Use pattern rules:

- Use the wildcard % as a placeholder in targets and dependencies.
- Use the special variable \$* to refer to matching sets of files in actions.
